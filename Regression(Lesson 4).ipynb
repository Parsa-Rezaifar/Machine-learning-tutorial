### First of all import needed packages

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Use this to show matplotlib and plots properly(But no essential)
%matplotlib inline
```

```python
# Make some random numbers from -5 to 5 by the distance of 0.1
x = np.arange(-5,5,0.1)
# Define a simple line(simple linear)
y = 2*(x)+3
# Normalize the data and double them(Normal distribution curve)
# This causes points are mostly close to the line ,also can be far from the line but it's less possible
# If it were'nt normal the distribution ratio could be equal
y_noise = 2 * np.random.normal(size=x.size)
y_data = y + y_noise
plt.plot(x,y_data,'bo')
plt.plot(x,y,'r')
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

### Let's see some non linear equations

#### Tertiary equation

```python
x = np.arange(-5,5,0.1)
y = 1*(x**3) + 1*(x**2) + 1*x +3
y_noise = 20 * np.random.normal(size=x.size)
y_data = y + y_noise
plt.plot(x,y_data,'bo')
plt.plot(x,y,'r')
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

#### Quadratic equation

```python
x = np.arange(-5,5,0.1)
y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
y_data = y + y_noise
plt.plot(x,y_data,'bo')
plt.plot(x,y,'r')
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

#### Exponential equation

```python
x = np.arange(-5,5,0.1)
y = np.exp(x)
plt.plot(x,y)
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

#### Logarithmic equation

```python
x = np.arange(-5,5,0.1)
y = np.log(x)
plt.plot(x,y)
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

#### Sigmoidal/Logistic(S-Curve)

```python
x = np.arange(-15,15,0.1)
y = 1-4/(1+np.power(3,x-2))
plt.plot(x,y)
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

```python
# The size of the figure 8 times 5
plt.figure(figsize=(8,5))
# Access to the values of each column and make them a simple array
x_data , y_data = (Data_Frame['Year'].values , Data_Frame['Value'].values)
plt.plot(x_data,y_data,'ro')
plt.xlabel('GDP')
plt.ylabel('Year')
plt.show()
```

```python
x = np.arange(-5,5,0.1)
y = 1/(1+np.exp(-x))
plt.plot(x,y)
plt.xlabel('Independent variable')
plt.ylabel('dependent variable')
plt.show()
```

```python
def sigmoid(x,Beta_1,Beta_2) :
    y = 1 / (1 + np.exp(-Beta_1 * (x-Beta_2)))
    return y
```

### Give sigmoid function some numbers just for testing

```python
beta_1 = 0.1
beta_2 = 1990
# Logistic function
y_prediction = sigmoid(x_data,beta_1,beta_2)
# Plot initial prediction against the datapoints
plt.plot(x_data,y_prediction*15000000000000)
plt.plot(x_data,y_data,'ro')
```

### Make the model a little more optimized(Manually)

```python
beta_1 = 0.8
# beta_2 is the slope
beta_2 = 2000
# Logistic function
y_prediction = sigmoid(x_data,beta_1,beta_2)
# Plot initial prediction against the datapoints
plt.plot(x_data,y_prediction*15000000000000)
plt.plot(x_data,y_data,'ro')
```

### What does curve_fit() do ?

```python
def my_function(x,a,b) :
    return a * x + b
# curve_fit(function,x_data,y_data)
x = [1,2,3,4]
y = [5,7,9,11]
# What is the best value for a and b to have those x and y exactly ?
curve_fit(my_function,x,y)
# If a is 2 and b is 3 there will be those x and y for us because the answers are all 0
# But curve_fit() also looks for the minimum error and the best co-efficients
```
