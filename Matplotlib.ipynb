# Let's see some basic usages of Matplotlib
### First of all import matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib
```

# let's do some basic stuff

```python
# This uses 1,2,3,4 for x axis and 1,2,3,4 for y axis
plt.plot([1,2,3,4])
plt.ylabel('Numbers')
plt.show()
```

```python
# plt.plot(x,y,'style')
plt.plot([1,3,4,9],[-1,8,1,2])
```

```python
# plt.plot(x,y,'style')
plt.plot([1,3,4,9],[-1,8,1,2],'ro')
```

```python
# plt.plot(x,y,'style')
plt.plot([1,3,4,9],[-1,8,1,2],'bo')
```

```python
# plt.plot(x,y,'style')
plt.plot([1,3,4,9],[-1,8,1,2],'b+')
```

```python
# plt.plot(x,y,'style')
t = np.arange(0,5,0.2)
plt.plot(t,t)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t*2)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t**2)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t**2,'r--')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r--')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r-')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r-.')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r-.',t,t**2,'bs')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r-.',t,t**2,'bs',t,t**3,'g^')
```
